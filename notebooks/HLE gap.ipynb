```python
# Import the tools we'll use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

# Settings to make output prettier
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Show charts in the notebook
%matplotlib inline

print("✓ Libraries loaded successfully!")
print(f"✓ Pandas version: {pd.__version__}")
```

```python
import os

# Check we can find our data folder
data_folder = '../data/'
print("Files in data folder:")
print(os.listdir(data_folder))
```

```python
# Load the Excel file to see what sheets it contains
hle_file = '../data/hle_2011_2024.xlsx'
excel_data = pd.ExcelFile(hle_file)

print("Sheets in the HLE file:")
for i, sheet in enumerate(excel_data.sheet_names, 1):
    print(f"{i}. {sheet}")
```

```python
# Let's look at what's in sheet "1"
df_test = pd.read_excel(hle_file, sheet_name='1')
print("Sheet 1 - First 10 rows:")
print(df_test.head(10))
print("\nColumn names:")
print(df_test.columns.tolist())
```

```python
# Load Sheet 1, skipping the first 5 rows to get to the real data
df_hle_raw = pd.read_excel(hle_file, sheet_name='1', skiprows=5)

print(f"✓ Loaded {len(df_hle_raw)} rows")
print("\nColumn names:")
print(df_hle_raw.columns.tolist())
print("\nFirst 5 rows:")
print(df_hle_raw.head())
```

```python
# Skip 6 rows total (not 5)
df_hle_raw = pd.read_excel(hle_file, sheet_name='1', skiprows=6)

print(f"✓ Loaded {len(df_hle_raw)} rows")
print("\nColumn names:")
print(df_hle_raw.columns.tolist())
print("\nFirst 5 rows:")
print(df_hle_raw.head())
```

```python
# See what time periods we have
print("Time periods in the data:")
print(sorted(df_hle_raw['Period'].unique()))

print("\n" + "="*50)

# See what sex categories we have
print("Sex categories:")
print(df_hle_raw['Sex'].unique())

print("\n" + "="*50)

# See what age groups we have
print("Age groups:")
print(df_hle_raw['Age group'].unique())

print("\n" + "="*50)

# See a few area names
print("Sample of area names (first 10 unique):")
print(df_hle_raw['Area name'].unique()[:10])
```

```python
# Let's check what's in Sheet 2 and Sheet 3
print("=== SHEET 2 ===")
df_sheet2 = pd.read_excel(hle_file, sheet_name='2', skiprows=6)
print("Columns:", df_sheet2.columns.tolist())
print("First few rows:")
print(df_sheet2.head(3))

print("\n=== SHEET 3 ===")
df_sheet3 = pd.read_excel(hle_file, sheet_name='3', skiprows=6)
print("Columns:", df_sheet3.columns.tolist())
print("First few rows:")
print(df_sheet3.head(3))
```

```python
# Check if there's an "at birth" value in the age groups
print("Checking for 'at birth' or 'birth' in age groups:")
age_groups = df_hle_raw['Age group'].unique()
for age in age_groups:
    if 'birth' in str(age).lower() or age == '0':
        print(f"Found: {age}")

# Also check what the <1 age group represents
print("\nSample data for '<1' age group:")
sample = df_hle_raw[df_hle_raw['Age group'] == '<1'].head(10)
print(sample[['Period', 'Area name', 'Sex', 'Age group', 'HLE']])
```

```python
# Filter to only "at birth" data (age group <1)
df_at_birth = df_hle_raw[df_hle_raw['Age group'] == '<1'].copy()

print(f"✓ Filtered to {len(df_at_birth)} rows (just 'at birth' data)")
print(f"\nWe now have:")
print(f"- {df_at_birth['Area name'].nunique()} unique local authorities")
print(f"- {df_at_birth['Period'].nunique()} time periods")
print(f"- {df_at_birth['Sex'].nunique()} sex categories")

print("\nSample of the data:")
print(df_at_birth.head(10))
```

```python
# Check what columns we have
print("Columns in our filtered data:")
print(df_at_birth.columns.tolist())

# Check if we have LE (Life Expectancy)
if 'LE' in df_at_birth.columns:
    print("\n✓ We have Life Expectancy data!")
    print("\nSample HLE vs LE:")
    print(df_at_birth[['Area name', 'Period', 'Sex', 'HLE', 'LE']].head(10))
else:
    print("\n✗ No LE column found. Need to find it.")
```

```python
# Look at the proportion column
print("Sample data with Proportion:")
print(df_at_birth[['Area name', 'Period', 'Sex', 'HLE', 'Proportion (%)']].head(10))

print("\n" + "="*50)

# If Proportion = (HLE / LE) × 100, then LE = HLE / (Proportion/100)
# Let's calculate it and see if it makes sense

df_at_birth['LE_calculated'] = df_at_birth['HLE'] / (df_at_birth['Proportion (%)'] / 100)

print("Calculated Life Expectancy:")
print(df_at_birth[['Area name', 'Period', 'Sex', 'HLE', 'Proportion (%)', 'LE_calculated']].head(10))

print("\n" + "="*50)
print("Does LE look reasonable? (should be ~75-85 years)")
print(f"Average calculated LE: {df_at_birth['LE_calculated'].mean():.1f} years")
```

```python
# Calculate Life Expectancy and HLE Gap
df_at_birth['LE'] = df_at_birth['HLE'] / (df_at_birth['Proportion (%)'] / 100)
df_at_birth['HLE_Gap'] = df_at_birth['LE'] - df_at_birth['HLE']

print("✓ Calculated LE and HLE Gap!")
print("\nSample data:")
print(df_at_birth[['Area name', 'Period', 'Sex', 'HLE', 'LE', 'HLE_Gap']].head(10))

print("\n" + "="*50)
print(f"Average HLE Gap: {df_at_birth['HLE_Gap'].mean():.1f} years")
print(f"Smallest gap: {df_at_birth['HLE_Gap'].min():.1f} years")
print(f"Largest gap: {df_at_birth['HLE_Gap'].max():.1f} years")
```

```python
# Save this cleaned data so you don't have to reload it every time
df_at_birth.to_csv('../outputs/hle_at_birth_cleaned.csv', index=False)

print("✓ Data saved to outputs/hle_at_birth_cleaned.csv")
print(f"✓ Total rows: {len(df_at_birth)}")
print(f"✓ Columns: {list(df_at_birth.columns)}")
print("\n🎉 DAY 3 COMPLETE! You now have clean, analysis-ready data.")
```
